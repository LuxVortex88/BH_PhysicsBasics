```python
import numpy as np
import matplotlib.pyplot as plt

def schwarzschild_radius(mass):
    G = 6.67430e-11
    c = 3.0e8
    return 2 * G * mass / c**2

mass_black_hole = 4e30  # 2 solar masses
rs = schwarzschild_radius(mass_black_hole)

r = np.linspace(rs * 1.1, rs * 10, 1000)
bending = 1.75 * (rs / r)  # Approximate light bending angle in radians

plt.plot(r / rs, bending)
plt.xlabel('Distance from center (r/rs)')
plt.ylabel('Light Bending (rad)')
plt.title('Gravitational Lensing Near a Black Hole')
plt.grid()
plt.show()
```
